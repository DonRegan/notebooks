# Power Spectrum

### Mean power spectrum $(\Delta k = 0.01,\,k_{\mathrm{min}}=0.005,\,k_{\mathrm{max}}=0.3)$

```python
# load data
Pz00 = loadtxt('./data/ZBOX2/PSpec/z0.0/kmax_0.3_dk_0.01_N512_avg.dat')
Pz05 = loadtxt('./data/ZBOX2/PSpec/z0.52/kmax_0.3_dk_0.01_N512_avg.dat')
Pmodel = loadtxt('./model/PSpec/1loop_Power_2222222.dat')

# compute growth factors
par = (0.25, 0.75, -1.0, 0.0)
gz00 = G(0.0, par)/G(0.0, par)
gz05 = G(0.52, par)/G(0.0, par)
gz10 = G(1.0, par)/G(0.0, par)

# one-loop predictions for the power spectrum
k = Pmodel[:,0]
Pmodel_z00 = gz00**2*Pmodel[:,1]+gz00**4*Pmodel[:,3]
Pmodel_z05 = gz05**2*Pmodel[:,1]+gz05**4*Pmodel[:,3]
Pmodel_z10 = gz10**2*Pmodel[:,1]+gz10**4*Pmodel[:,3]

# plotting
f = plt.figure(figsize=(9,8))
ax = f.add_subplot(111)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('$k$ [h Mpc$^{-1}$]',fontsize=16)
ax.set_ylabel('$P$ [(Mpc/h)$^3$]',fontsize=16)
ax.set_xlim(0.001,0.5)

ax.errorbar(Pz00[:,0],Pz00[:,1],Pz00[:,2],fmt='x',mew=2,lw=2,c='b',label='z = 0.0')
ax.errorbar(Pz00[:,0],Pz05[:,1],Pz05[:,2],fmt='+',mew=2,lw=2,c='r',label='z = 0.52')

ax.plot(k, Pmodel_z00, lw=2, c='b')
ax.plot(k, Pmodel_z05, lw=2, c='r')
ax.plot(k, Pmodel_z10, lw=2, c='black')

legend(loc=1,prop={'size':15},frameon=False)
```

# Bispectrum

### Mean bispectrum $(\Delta k = 8 k_f,\,k_{\mathrm{min}}=4 k_f,\,k_{\mathrm{max}}=0.3)$

```python
# tree-level prediction
Pmodel = loadtxt('./model/PSpec/1loop_Power_2222222.dat')
Pspline = interp1d(Pmodel[:,0],Pmodel[:,1],kind='cubic')

par = (0.25, 0.75, -1.0, 0.0)
gz00 = G(0.0, par)/G(0.0, par)
gz05 = G(0.52, par)/G(0.0, par)
gz10 = G(1.0, par)/G(0.0, par)

def F2(k1, k2, k3):
    mu = (k3**2-k1**2-k2**2)/(2*k1*k2)
    return 5./7 + mu/2*(k1/k2+k2/k1) + 2./7*mu**2

nshapes = 95
dk = 8*2*pi/1500
k1 = dk/2
Bmodel_z00 = zeros(nshapes)
Bmodel_z05 = zeros(nshapes)
Bmodel_z10 = zeros(nshapes)
i = 0
while k1 < 0.3:
    k2 = k1
    while k2 < 0.3:
        k3 = k2
        while k3 < 0.3:
            if k3 < k1+k2:
                B = 2*(F2(k1,k2,k3)*Pspline(k1)*Pspline(k2)+F2(k2,k3,k1)*Pspline(k2)*Pspline(k3)+F2(k3,k1,k2)*Pspline(k3)*Pspline(k1))
                Bmodel_z00[i] = gz00**4*B
                Bmodel_z05[i] = gz05**4*B
                Bmodel_z10[i] = gz10**4*B
                i += 1
            k3 += dk
        k2 += dk
    k1 += dk
```

```python
# load data
nshapes = 95
Bz00 = loadtxt('./data/ZBOX2/BSpec/z0.0/kmax_0.3_dk_8kf_N512_avg.dat')
Bz05 = loadtxt('./data/ZBOX2/BSpec/z0.52/kmax_0.3_dk_8kf_N512_avg.dat')

# plotting
f = plt.figure(figsize=(9,8))
ax = f.add_subplot(111)

ax.set_yscale('log')
ax.set_xlabel('Triangle Index',fontsize=16)
ax.set_ylabel('$B$ [(Mpc/h)$^6$]',fontsize=16)
ax.set_xlim(-1,96)

ax.errorbar(arange(0,nshapes),Bz00[:,3],Bz00[:,4],fmt='x',mew=2,lw=2,c='b',label='z = 0.0')
ax.errorbar(arange(0,nshapes),Bz05[:,3],Bz05[:,4],fmt='+',mew=2,lw=2,c='r',label='z = 0.52')

ax.plot(arange(0,nshapes),Bmodel_z00,lw=2,c='b')
ax.plot(arange(0,nshapes),Bmodel_z05,lw=2,c='r')
ax.plot(arange(0,nshapes),Bmodel_z10,lw=2,c='black')

legend(loc=1,prop={'size':15},frameon=False)
```

### Correlation matrix $(z=0.0)$

```python
# power spectrum bins
k = 0.005+0.01*arange(0,30)

# triangle indices
tri = arange(0,nshapes)

# correlation matrix (including cross-covariance)
R_PB_z00 = correlation(0.0, ['PSpec','BSpec'], [k,tri])

# plotting
f = plt.figure(figsize=(9,8))
ax = f.add_subplot(111)

ax.set_xlim(-0.5,124.5)
ax.set_ylim(-0.5,124.5)

cax = ax.imshow(flipud(R_PB_z00), extent=[-0.5,124.5,-0.5,124.5], aspect='equal', interpolation='nearest',
                cmap='RdGy', vmin=-1, vmax=1)
                #cmap=shiftedColorMap(matplotlib.cm.RdGy, midpoint=1-amax(Rz00)/(amax(Rz00)+absolute(amin(Rz00)))))
    
ax.plot((29.5,29.5),(-0.5,124.5),'black')
ax.plot((-0.5,124.5),(29.5,29.5),'black')
                
cb = f.colorbar(cax)
```
