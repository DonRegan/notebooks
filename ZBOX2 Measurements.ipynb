# Bispectrum

### Mean bispectrum (averaged over 200 realizations)

```python
# load data
nshapes = 95
Bz00 = loadtxt('./data/ZBOX2/BSpec/z0.0/kmax_0.3_dk_8kf_N512_avg.dat')
Bz05 = loadtxt('./data/ZBOX2/BSpec/z0.52/kmax_0.3_dk_8kf_N512_avg.dat')

f = plt.figure(figsize=(9,8))
ax = f.add_subplot(111)

ax.set_yscale('log')
ax.set_xlabel('Triangle Index',fontsize=16)
ax.set_ylabel('$B$ [(Mpc/h)$^6$]',fontsize=16)
ax.set_xlim(-1,96)

ax.errorbar(arange(0,nshapes),Bz00[:,3],Bz00[:,4]/sqrt(197),fmt='-x',mew=2,lw=2,c='b',label='z = 0.0')
ax.errorbar(arange(0,nshapes),Bz05[:,3],Bz05[:,4]/sqrt(198),fmt='-+',mew=2,lw=2,c='r',label='z = 0.52')

legend(loc=1,prop={'size':15},frameon=False)
```

# Integrated Bispectrum

### Mean integrated bispectrum (for 5 subvolumes)

```python
# load data
iBz00 = loadtxt('./data/ZBOX2/iBSpec/z0.0/nCut_5_kmax_0.3_dk_0.01_N512_avg.dat')

f = plt.figure(figsize=(9,8))
ax = f.add_subplot(111)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('$k$ [h Mpc$^{-1}$]',fontsize=16)
ax.set_ylabel('$iB(k)$ [(Mpc/h)$^6$]',fontsize=16)

ax.errorbar(iBz00[:,0],iBz00[:,1],iBz00[:,2]/sqrt(197),fmt='x',mew=2,lw=2,c='b',label='z = 0.0')

legend(loc=1,prop={'size':15},frameon=False)
```
